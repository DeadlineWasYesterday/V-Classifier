```python
import pandas as pd
```

```python
df2 = pd.concat([pd.read_csv('../ppnnw5.csv'), pd.read_csv('../ppnnw6.csv').iloc[:,10:]], axis = 1)
```

```python
p = df2.iloc[:2862, :]
pi = df2.iloc[2862:5724, :]
n = df2.iloc[5724:6007, :]
n2 = df2.iloc[6007:, :]
```

```python
df = pd.DataFrame(index= n.iloc[:,10:].columns, columns = [0,1,2,3]).fillna(0)


def whatever(p,pi,n,n2):    
    for c in p.iloc[:,10:].columns:
        r11 = min(p[c])
        r12 = max(p[c])
        r21 = min(pi[c])
        r22 = max(pi[c])
        r31 = min(n[c])
        r32 = max(n[c])
        r41 = min(n2[c])
        r42 = max(n2[c])

        for number in p[c].to_list():
            c1 = 0
            if r21 < number < r22:
                c1+=1
            if r31 < number < r32:
                c1+=1
            if r41 < number < r42:
                c1+=1

            df.loc[c,c1] = df.loc[c,c1] + 1

        for number in pi[c].to_list():
            c1 = 0
            if r11 < number < r12:
                c1+=1
            if r31 < number < r32:
                c1+=1
            if r41 < number < r42:
                c1+=1

            df.loc[c,c1] = df.loc[c,c1] + 1

        for number in n[c].to_list():
            c1 = 0
            if r11 < number < r12:
                c1+=1
            if r21 < number < r22:
                c1+=1
            if r41 < number < r42:
                c1+=1

            df.loc[c,c1] = df.loc[c,c1] + 1

        for number in n2[c].to_list():
            c1 = 0
            if r11 < number < r12:
                c1+=1
            if r21 < number < r22:
                c1+=1
            if r31 < number < r32:
                c1+=1

            df.loc[c,c1] = df.loc[c,c1] + 1
            
            
for i in range(50):
    a = p.loc[p['die'] == 'no'].sample(240)
    b = pi.loc[pi['die'] == 'no'].sample(240)
    c = n.loc[n['die'] == 'no'].sample(240)
    d = n2.loc[n2['die'] == 'no'].sample(240)
    
    whatever(a,b,c,d)
    print('%d remaining.' %(50 - i))
```
