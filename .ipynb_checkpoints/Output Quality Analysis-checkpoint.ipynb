```python
import pandas as pd

data = pd.read_csv('proteinGroups.txt', sep = '\t', header = None)

pd.options.display.max_columns = None
```

```python
display(data.head())
```

```python
#Read MaxQuant params file

import xml.etree.ElementTree as ET
root = ET.parse('mqpar1.xml').getroot()
```

```python
#Parse experimental design from mqpar

l = [] 

for n in root.iter('experiments'):
    for n2 in n.findall('string'):
        l.append(n2.text)

experiments = len(set(l))
```

```python
#LFQ column names

expected_columns = []

for i in range(experiments):
    expected_columns.append('LFQ intensity ' + str(i+1))
```

```python
#setting columns and dropping duplicate columns row

data.columns = data.iloc[0]
data = data.drop(labels = 0, axis = 0).reset_index(drop = True)
```

```python
#new df out of only LFQ columns

df = data.loc[: , expected_columns[0]:expected_columns[-1]]
```

```python
#Assigning tissue type and source ID to LFQ Intensities

exp_list = list(range(1, experiments + 1))

#Assigning variables to tissue names

tis1 = 'Andrenalgland'
tis2 = 'Colon'
tis3 = 'Kidney'
tis4 = 'Liver'
tis5 = 'Lung'
tis6 = 'Prostate'
tis7 = 'Retina'
tis8 = 'Spinalcord'

#Assigning sources

sources = {"PXD000561" : [tis1, tis1, tis1, tis5, tis5, tis5, tis2, tis2]}

# sources = {"PXD000561" : [tis1, tis1, tis1, tis2, tis2, tis5, tis5, tis5, tis6, tis6, tis7, tis7, tis7, tis8, tis8] ...
#            , "PXD010154" : [tis1, tis2, tis5, tis6] ...
#           }

#Filling in column names

column_names = []

for k,v in sources.items():
    c = 1
    for t in v:
        column_names.append(k + ' ' + t + ' '  + 'LFQ intensity ' + str(c))
        c+=1
```
