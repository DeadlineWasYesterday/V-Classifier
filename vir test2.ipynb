```python
from Bio import SeqIO


eyed = []
name = []
desc = []

for seq_record in SeqIO.parse('Complete NCBI Virus Genbank only.fasta', "fasta"):
    eyed.append(seq_record.id)
    desc.append(seq_record.description)
```

```python
#Hardcode anomaly
# desc.index('MK293717 |Gemycircularvirus strain RN/Haikou/China| complete genome|complete|China|Rattus norvegicus|') gives output 21244

desc[21244] = 'MK293717 |Gemycircularvirus strain RN/Haikou/China| complete genome|complete|China|Rattus norvegicus'
```

```python
df = pd.DataFrame({'Sequence description' : desc})
df['ID'] = eyed
```

```python
l = [x.split('|')[-1].lower() for x in df['Sequence description']]
df['Species'] = l
```

```python
df['Genome'] = 'dunno'
```

```python
df2 = df.drop_duplicates(subset = 'Species')
```

```python
# Get rid of numbers or any word that has numbers from Species

w = []
tl = list(df2['Species'])

for i in df2['Species']:
    for i2 in i.split(' '):
        for i3 in i2:
            if i3.isdigit():
                w.append((i2, i, tl.index(i)))
```
